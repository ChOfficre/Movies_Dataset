```python
df_best = movies_df[["poster_path", "title", "budget_musd", "revenue_musd", "vote_count", "vote_average", "popularity"]].copy()
df_best
```

```python
df_best["profit_musd"] = movies_df.revenue_musd.sub(movies_df.budget_musd)
df_best["return"] = movies_df.revenue_musd.div(movies_df.budget_musd)
df_best
```

```python
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes",  "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
df_best.set_index("Title", inplace=True)
df_best
```

Filtering Movis DF by criteria

```python
df_best.loc[df_best.Budget >= 5].sort_values(by='ROI', ascending=False)
```

Fill missing values with 0

```python
df_best["Budget"] = df_best["Budget"].fillna(0)
df_best["Votes"] = df_best["Votes"].fillna(0)
df_best.info()
```

```python
def best_worst(n, by, ascending=False, min_bud=0, min_votes=0):

    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes), ["", by ]].sort_values(by=by, ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

Top 5 movies by Highest Revenueb

```python
best_worst(n=5, by="Revenue")
```

Top 5 movies by Highest Budget

```python
best_worst(n=5, by="Budget")
```

Top 5 movies by Highest Profitb

```python
best_worst(n=5, by="Profit")
```

Top 5 movies by Lowest Profit

```python
best_worst(n=5, by="Profit", ascending=True)
```

Top 5 movies by Highest ROI

```python
best_worst(n=5, by="ROI", min_bud=5)
```

Top 5 movies by Lowest ROI

```python
best_worst(n=5, by="ROI", ascending=True, min_bud=100)
```

Top 5 movies by Most Votes

```python
best_worst(n=5, by="Votes")
```

Top 5 movies by Highest Rating

```python
best_worst(n=5, by="Average Rating", min_votes=20)
```

Top 5 movies by Most Popular

```python
best_worst(n=5, by="Popularity", min_bud=10, min_votes=20)
```

# Are Franchises more Successful?

```python
movies_df.belongs_to_collection
```

```python
movies_df["Franchise"] = movies_df.belongs_to_collection.notna()
movies_df.Franchise
```

```python
movies_df.Franchise.value_counts()
```

Franchies vs. Movies. Average Revenus

```python
movies_df.groupby("Franchise").revenue_musd.mean()
```

Franchies vs. Movies. Return on Investment / Profitability

```python
movies_df["ROI"] = movies_df.revenue_musd / movies_df.budget_musd
```

```python
movies_df.groupby("Franchise").ROI.median()
```

```python
movies_df.groupby("Franchise").agg({"budget_musd": "mean", "revenue_musd": "mean", "vote_average": "mean", "popularity": "mean", "ROI": "median", "vote_count": "mean"})
```

Most Successful Franchises

```python
movies_df.belongs_to_collection
```

```python
movies_df.belongs_to_collection.value_counts()
```

```python
franchise = movies_df.groupby("belongs_to_collection").agg({"title": "count", "budget_musd": ["sum", "mean"], "revenue_musd": ["sum", "mean"], "vote_average": "mean", "popularity": "mean", "ROI": "median", "vote_count": "mean"})
franchise
```

```python
franchise.nlargest(20, ("title", "count"))
```
