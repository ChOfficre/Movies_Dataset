Filtering Movis DF by criteria

```python
df_best.loc[df_best.Budget >= 5].sort_values(by='ROI', ascending=False)
```

Fill missing values with 0

```python
df_best["Budget"] = df_best["Budget"].fillna(0)
df_best["Votes"] = df_best["Votes"].fillna(0)
df_best.info()
```

```python
def best_worst(n, by, ascending=False, min_bud=0, min_votes=0):

    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes), ["", by ]].sort_values(by=by, ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

Top 5 movies by Highest Revenueb

```python
best_worst(n=5, by="Revenue")
```

Top 5 movies by Highest Budget

```python
best_worst(n=5, by="Budget")
```

Top 5 movies by Highest Profitb

```python
best_worst(n=5, by="Profit")
```

Top 5 movies by Lowest Profit

```python
best_worst(n=5, by="Profit", ascending=True)
```

Top 5 movies by Highest ROI

```python
best_worst(n=5, by="ROI", min_bud=5)
```

Top 5 movies by Lowest ROI

```python
best_worst(n=5, by="ROI", ascending=True, min_bud=100)
```

Top 5 movies by Most Votes

```python
best_worst(n=5, by="Votes")
```

Top 5 movies by Highest Rating

```python
best_worst(n=5, by="Average Rating", min_votes=20)
```

Top 5 movies by Most Popular

```python
best_worst(n=5, by="Popularity", min_bud=10, min_votes=20)
```

# Most successful/popular genres

```python
movies_df.genres.value_counts()
```

```python
genres = movies_df.genres.str.split("|", expand=True)
genres = genres.stack().reset_index(level=1, drop=True).to_frame()
genres
```

```python
genres.columns = ["Genres"]
gen = genres.merge(movies_df[["title","release_date", "production_companies", "revenue_musd", "vote_average", "popularity"]], how="left", left_index=True, right_index=True)
gen
```

```python
plt.figure(figsize=(10, 5))
movies_df.genres.value_counts().head(20).plot(kind="bar", fontsize=15)
plt.title("Genres Count", fontsize=15)
plt.ylabel("Number of Genres", fontsize=10)
plt.show
```

```python
genres_agg = gen.groupby(["Genres", "title", "production_companies"]).agg(Release_Date = ("release_date", "mean"),
                                      Total_Revenue = ("revenue_musd", sum),
                                      Mean_Revenue = ("revenue_musd", "mean"),
                                      Mean_rating = ("vote_average", "mean"),
                                      Popularity = ("popularity", "mean"))
```

```python
genres_agg.nlargest(20, "Total_Revenue")
```

```python
popularity_of_80_90 = genres_agg.loc[genres_agg["Release_Date"].between("1980-01-01", "1999-12-31")].nlargest(20, "Popularity")
popularity_of_80_90
```

Highest-rated Action movies of the 90ths

```python
gen_action = gen[gen.Genres == "Action"]
gen_action_90 = gen_action.loc[gen_action["release_date"].between("1990-01-01", "1999-12-31")].nlargest(20, "popularity")
gen_action_90
```
